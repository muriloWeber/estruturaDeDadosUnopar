```python
# Classe ItemLista

class ItemLista:

  # representa cada item de uma lista encadeada

  def __init__(self, data=0, nextItem=None):
    self.data = data
    self.nextItem = nextItem
  def __repr__(self):
    return '%s => %s' % (self.data, self.nextItem)
  
```

O método __init__ instancia a classe e determina dois atributos necessários: data e nextItem. O segundo é __repr__ que retorna uma mensagem.

```python
class ListaEncadeada:

  # cria uma lista encadeada

  def __init__(self):

    self.head = None

  def __repr__(self):

    return "%s" % (self.head)

  def insere(lista, data):

  # cria um objeto para armazenar um novo item da lista
    item = ItemLista(data)
  # o head é apontado como próximo item
    item.nextItem = lista.head
  # o item atual se torna o head
    lista.head = item

  def busca(lista, valor):
    navegar = lista.head
    while navegar and navegar.data != valor:
      navegar = navegar.nextItem
    return navegar

  def remove(self, valor):
    # verificamos se o item se trata do head
    if self.head.data == valor:
      self.head = self.head.nextItem
    else:
      # detectando a posição do elemento
      before = None
      navegar = self.head
      # navega pelo cabeçalho atual
      while navegar and navegar.data != valor:
        before = navegar
        navegar = navegar.nextItem
        # remove o item atual
      if navegar:
        before.nextItem = navegar.nextItem
      else:
        before.nextItem = None
```

Aqui, esses métodos possuem os mesmos objetivos que na classe ItemLista. Ao instanciar essa classe, observe que apenas um objeto head é determinado.

```python
# cria o objeto

lista = ListaEncadeada()
print("Conteúdo da lista:", lista) # lista está vazia
```

```python
# vamos adicionar alguns itens na lista:

# cria a lista

ListaEncadeada.insere(lista, "abacate")
ListaEncadeada.insere(lista, "biscoito")
ListaEncadeada.insere(lista, "cenoura")
ListaEncadeada.insere(lista, "desodorante")
ListaEncadeada.insere(lista, "espinafre")
print(lista)
```
